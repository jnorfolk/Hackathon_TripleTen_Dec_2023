```python
# function to categorize products
def categorize_product(product):
    product_lower = product.lower()
    if any(keyword in product_lower for keyword in ['loan', 'mortgage']):
        return 'loans'
    elif any(keyword in product_lower for keyword in ['money transfer', 'currency', 'money service']):
        return 'money transfer'
    else:
        return 'other'

# appling the function to the 'Product' column to create a new 'product_type' column
complaints['product_type'] = complaints['product'].apply(categorize_product)

# checking the result
complaints.head(1)
```

```python
# creating new table with loans and money transfers
types_of_product = ['loans', 'money transfer']
df = complaints.query('product_type in @types_of_product').reset_index(drop = True)
df.head(1)
```
